<h2 style="margin:0;"><b>3) Diagnostics A — feature set × k silhouette grid</b></h2>
<p style="margin:6px 0 0 0;">A single heatmap of mean silhouette for every feature set across k, under one scaler. Higher = better separated. Re-run with a different <code>scaler_name</code> to see how sensitive things are.</p>

<div style="background-color:#F7FBFF; padding:15px; border-radius:10px;">
<b>How to read it.</b> Look for a row (feature set) that stays high, and a column (k) where it peaks. A high silhouette means tight, well-separated clusters — but a very high score at k = 2 usually just means "everyone vs a tiny group", which is rarely useful for marketing. Balance separation against how many distinct, targetable segments you actually want.
</div>

### 3.1 — Principled feature-set search (granular combos + embedded ranking)

<div style="background-color:#F7FBFF; padding:15px; border-radius:10px;">
Instead of dumping every variable into the distance, we compare <b>curated</b> combinations &mdash; spend, then one behaviour/lifecycle block at a time. We also cluster on a broad set and read the <b>embedded</b> importance to see which features the segmentation actually uses; features with importance ~0 (e.g. <code>is_male</code>) are dropped. Silhouette favours fewer features, so when two sets score similarly we keep the one whose surviving features are more useful for marketing (e.g. promo sensitivity).
</div>

```python
granular = uc.build_granular_feature_sets(regular)
gg = uc.silhouette_grid(regular, granular, k_range=range(7, 11), scaler_name="Standard")
display(gg.pivot(index="feature_set", columns="k", values="silhouette"))

# Which features does a BROAD solution actually rely on? Near-zero -> drop.
broad_cols = (granular["spend_ng + promo + engagement + family"][0]
              + [c for c in ["customer_age", "tenure", "education_level",
                             "number_complaints", "is_male", "customer_loyalty_flag"]
                 if c in regular.columns])
ranking = uc.rank_features_for_clustering(regular, broad_cols, k=8, scaler_name="Standard")
display(ranking)
print("Drop candidates (forest importance < 0.02):",
      ranking.index[ranking["forest_importance"] < 0.02].tolist())
```

<div style="background-color:#FFF6E5; padding:20px; border-radius:10px; border-left:8px solid #E08600;">
<h2 style="margin:0;"><b>5) YOUR CHOICES</b></h2>
<p style="margin:6px 0 0 0;">This is the only cell that makes a decision. Set the three values below from the menus above, then run everything that follows. Change them and re-run as often as you want.</p>
</div>

```python
# ===========================  YOUR DECISIONS  ===========================
# 1) FEATURE_SET : any key printed in Section 2
# 2) SCALER      : 'Standard' | 'MinMax' | 'Robust' | 'None'
# 3) K           : the number of clusters you decide on
# ------------------------------------------------------------------------
FEATURE_SET = 'lifetime_spend no groceries'  # silhouette grid + embedded check favour this
SCALER      = 'Robust'                        # confirm with the Section 4 comparison (now aligned)
K           = 8                               # inside the 7-10 range; elbow + interpretability
# ========================================================================

distance_cols, LOGABS = candidate_sets[FEATURE_SET]
profiling_cols = uc.get_profiling_features(regular, distance_cols)
scaler = uc.get_scaler(SCALER)
X = uc.apply_feature_pipeline(regular, distance_cols, LOGABS, scaler, fit=True)

print(f'Feature set : {FEATURE_SET}  ({len(distance_cols)} cols, log_abs={LOGABS})')
print(f'Scaler      : {SCALER}')
print(f'k           : {K}')
print(f'Matrix      : {X.shape}')
```

<div style="background-color:#FFF6E5; padding:16px; border-radius:10px; border-left:8px solid #E08600;">
<b>Decision log.</b> We cluster on <b>absolute lifetime spend excluding groceries</b> (<code>lifetime_spend no groceries</code>). In the Section&nbsp;3 silhouette grid this set scores consistently above the version that includes groceries, and groceries &mdash; which almost every customer buys heavily &mdash; otherwise dominates the Euclidean distance and washes out the category differences that make segments actionable. The embedded check in Section&nbsp;6.3 independently ranks groceries among the <i>least</i> important features for separating the segments, which confirms the choice. Groceries stay in the data and are used in the profiling tables.<br><br>We use the scaler that wins the Section&nbsp;4 comparison on this exact feature set, and fix <b>k&nbsp;=&nbsp;8</b>: it lies inside the requested 7&ndash;10 range, the inertia elbow has flattened by then, and it yields distinct, sizeable, namable personas. We accept a moderate silhouette (~0.2), which is normal for high-dimensional behavioural data, in exchange for segments that are interpretable and large enough to target.
</div>

### 6.3 — Embedded feature-importance check (validates the feature set)

<div style="background-color:#F7FBFF; padding:15px; border-radius:10px;">
The correlation matrix in Notebook&nbsp;1 is a <b>filter</b> method: it only removes <i>redundant</i> features and cannot say which features actually <i>drive</i> the segments. Here we add an <b>embedded</b> method &mdash; an L1-penalised (Lasso) logistic model and a Random Forest &mdash; using the cluster label as a pseudo-target. The L1 penalty shrinks uninformative coefficients to exactly zero (an embedded selection), and both models rank how much each feature contributes to telling the segments apart. A peaked, sparse ranking means the segmentation rests on a few interpretable drivers. This is a post-hoc validation of the feature set chosen in Section&nbsp;5, not a pre-selection step.
</div>

```python
emb_importance = uc.embedded_feature_importance(
    X, kmeans.labels_, distance_cols, method='both', random_state=0
)
display(emb_importance)
uc.plot_embedded_importance(emb_importance,
                            title=f'Embedded feature importance - {FEATURE_SET}')

print('Embedded shortlist (forest importance > median):',
      uc.select_features_embedded(emb_importance, 'forest_importance'))
```

### 6.4 — Ensemble / consensus clustering (robustness check)

<div style="background-color:#F7FBFF; padding:15px; border-radius:10px;">
A single K-Means depends on its initialisation. We run it many times, align the labels across runs (Hungarian algorithm on the confusion matrix) and majority-vote each customer. This gives a <b>consensus</b> segmentation and a per-customer <b>stability</b> score (fraction of runs that agree). High stability means the segments are real, not a lucky seed; low-stability customers sit between segments. The stability can be exported as a confidence column alongside each customer’s cluster.
</div>

```python
consensus_labels, stability = uc.consensus_kmeans(X, K, n_runs=25, random_state=0)
regular["cluster_stability"] = stability

from sklearn.metrics import silhouette_score, adjusted_rand_score
print("Single-run silhouette  :", round(silhouette_score(
    X, kmeans.labels_, sample_size=8000, random_state=0), 4))
print("Consensus  silhouette  :", round(silhouette_score(
    X, consensus_labels, sample_size=8000, random_state=0), 4))
print("Agreement single vs consensus (ARI):",
      round(adjusted_rand_score(kmeans.labels_, consensus_labels), 4))
print("Mean stability:", round(stability.mean(), 3),
      "| very stable (>=0.9):", f"{(stability>=0.9).mean()*100:.1f}%",
      "| ambiguous (<0.6):", f"{(stability<0.6).mean()*100:.1f}%")

uc.plot_stability(stability)
```

<div style="background-color:#EFF5FF; padding:20px; border-radius:10px; border-left:8px solid #0047AB;">
<h2 style="margin:0;"><b>7) Hierarchical cross-check</b></h2>
<p style="margin:6px 0 0 0;">Compare your K-Means labels with a Ward solution at the same k on a sample. Strong agreement (most mass on one Ward column per K-Means row) means the structure is real.</p>

```python
sample_idx, ward_labels = uc.fit_hierarchical_sample(X, K, sample_size=5000, linkage='ward')
uc.compare_solutions(kmeans.labels_[sample_idx], ward_labels, name_a='KMeans', name_b='Ward')
```

### 7.2 — Hierarchical on KMeans centroids

This applies a centroid-based hierarchical strategy using the current feature set. First we create 20 KMeans micro-clusters, then apply Ward hierarchical clustering to the 20 centroids and cut the dendrogram into `K` macro-clusters. This is a benchmark; the final export below still uses the KMeans solution selected in Section 5 unless you decide otherwise.


```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.metrics import silhouette_score

MICRO_K = 20
micro_kmeans = uc.fit_kmeans(X, MICRO_K, random_state=0, n_init=20)
regular["micro_cluster"] = micro_kmeans.labels_

centroids_df = pd.DataFrame(
    micro_kmeans.cluster_centers_,
    columns=distance_cols,
    index=[f"C{i}" for i in range(MICRO_K)],
)

Z = linkage(centroids_df.values, method="ward")

plt.figure(figsize=(10, 5))
dendrogram(
    Z,
    labels=centroids_df.index.tolist(),
    leaf_rotation=90,
    leaf_font_size=10,
)
plt.title("Hierarchical clustering on KMeans centroids (Ward linkage)")
plt.ylabel("Ward distance")
plt.tight_layout()
plt.show()

macro_labels_for_centroids = fcluster(Z, t=K, criterion="maxclust") - 1
centroid_to_macro = dict(zip(range(MICRO_K), macro_labels_for_centroids))
regular["macro_cluster"] = regular["micro_cluster"].map(centroid_to_macro)

print("Micro-cluster sizes:")
display(regular["micro_cluster"].value_counts().sort_index())

print("Macro-cluster sizes:")
display(regular["macro_cluster"].value_counts().sort_index())

macro_sil = silhouette_score(X, regular["macro_cluster"].values)
print(f"Macro-cluster silhouette: {macro_sil:.4f}")
print(f"Final KMeans silhouette: {avg_sil:.4f}")
```

```python
macro_comparison = uc.compare_solutions(
    regular["cluster"].values,
    regular["macro_cluster"].values,
    name_a="Final KMeans",
    name_b="Centroid-Ward macro",
)
display(macro_comparison)

uc.plot_silhouette_blades(
    X,
    regular["macro_cluster"].values,
    title=f"Silhouette — Hierarchical on KMeans centroids (k={K})",
)
```

```python
X_macro_emb, macro_lab_emb = uc.subsample(X, regular["macro_cluster"].values, n=8000)
macro_pca_emb = uc.embed_pca(X_macro_emb)
uc.plot_embedding(
    macro_pca_emb,
    macro_lab_emb,
    title="PCA - Centroid-Ward Macro Clusters",
    method_name="PCA",
)

macro_umap_emb, macro_method = uc.embed_umap(X_macro_emb)
uc.plot_embedding(
    macro_umap_emb,
    macro_lab_emb,
    title="UMAP - Centroid-Ward Macro Clusters",
    method_name=macro_method,
)
```
