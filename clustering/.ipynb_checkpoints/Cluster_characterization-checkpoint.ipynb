# Cluster Characterization

This notebook characterizes the final customer segments after the clustering model has been selected. The objective is to translate numerical cluster labels into interpretable customer profiles using spend, behavioural, demographic, complaint and geographic variables.


## 1) Imports and data loading


```python
%load_ext autoreload
%autoreload 2

import os, sys
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sys.path.append(os.path.abspath('.'))
import utils_cluster_characterization as ucc

sns.set_theme(style="whitegrid")
pd.set_option("display.max_columns", None)

DATA_DIR = "../datasets"
df = ucc.load_characterization_data(DATA_DIR)
print("Characterization dataset shape:", df.shape)
df.head()
```

## 2) Segment sizes

The first check is the size of each segment. Very small clusters would require caution because they may represent niche behaviour or residual outlier structure rather than stable customer segments.


```python
size_profile = ucc.cluster_sizes(df)
display(size_profile)
ucc.plot_cluster_sizes(size_profile)
```

## 3) Spend profile

Absolute lifetime spend by category is used to understand the main purchasing orientation of each segment. These variables are kept in original units for interpretability.


```python
spend_cols = [c for c in df.columns if c.startswith("lifetime_spend_")]
spend_profile = ucc.profile_table(df, spend_cols)
display(spend_profile)
ucc.plot_profile_heatmap(spend_profile, title="Lifetime spend profile by cluster")
```

## 4) Behavioural and demographic profile

This table adds non-spend variables such as promotion usage, product diversity, age, education, tenure, children and complaints. These features are useful for assigning business names to clusters.


```python
profile_cols = [c for c in [
    "log_total_spend",
    "percentage_of_products_bought_promotion",
    "distinct_stores_visited",
    "lifetime_total_distinct_products",
    "customer_age",
    "education_level",
    "tenure",
    "total_children",
    "kids_home",
    "teens_home",
    "number_complaints",
    "customer_loyalty_flag",
    "is_male",
    "typical_hour",
] if c in df.columns]

behaviour_profile = ucc.profile_table(df, profile_cols)
display(behaviour_profile)
ucc.plot_profile_heatmap(behaviour_profile, title="Behavioural and demographic profile")
```

## 5) Normalised comparison

Because the variables have different units, a min-max scaled comparison is used to identify which clusters are highest or lowest on each feature.


```python
combined_profile = ucc.profile_table(df, spend_cols + profile_cols)
scaled_profile = ucc.plot_scaled_profile(combined_profile)
display(scaled_profile)
```

## 6) Feature-by-feature plots

The barplots and boxplots check whether the differences observed in the profile tables are clear across the distribution or mainly driven by extreme values.


```python
key_plot_cols = [c for c in [
    "lifetime_spend_groceries",
    "lifetime_spend_vegetables",
    "lifetime_spend_alcohol_drinks",
    "lifetime_spend_meat",
    "lifetime_spend_hygiene",
    "lifetime_spend_petfood",
    "lifetime_spend_technology",
    "percentage_of_products_bought_promotion",
    "customer_age",
    "total_children",
    "number_complaints",
] if c in df.columns]

ucc.plot_feature_bars(df, key_plot_cols)
ucc.plot_boxplot_grid(df, [
    "log_total_spend",
    "percentage_of_products_bought_promotion",
    "customer_age",
    "total_children",
    "number_complaints",
])
```

## 7) Main differentiators by cluster

The following table lists the largest relative differences from the overall customer base. It is used as a compact evidence table for cluster naming.


```python
deviations = ucc.top_deviations(combined_profile, n=6)
display(deviations)
```

## 8) Geographic check

Latitude and longitude are used only for profiling. This plot checks whether any final segment is associated with a specific geographic concentration.


```python
if {"latitude", "longitude"}.issubset(df.columns):
    plt.figure(figsize=(8, 7))
    sns.scatterplot(data=df, x="longitude", y="latitude", hue="cluster", palette="tab10", s=8, alpha=0.45)
    plt.title("Customer locations by segment")
    plt.tight_layout()
    plt.show()
```

## 9) Final segment names

The names below are descriptive labels based on the profiling evidence. They can be adjusted if the final clustering configuration changes.


```python
display(
    df[["cluster", "cluster_name"]]
    .drop_duplicates()
    .sort_values("cluster")
    .reset_index(drop=True)
)
```

## 10) Export ID and cluster mapping


```python
id_cluster = ucc.export_id_cluster(df, f"{DATA_DIR}/id_and_cluster.csv")
print("Exported rows:", len(id_cluster))
id_cluster.head()
```
